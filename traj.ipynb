```python
read_dframe = pd.read_csv('create_dframe.csv')
```

```python
xaxis_person1 =[]
count = 0
yaxis_person1 =[]
for column in read_dframe.loc[:,'bounding_boxes']:
    # Iterate over column names
    x = eval(column)
    get_current_row =0
    count = 0
    for i in x:
        if(count == 0):
            centreRect1x = (i[0] + i[2])/2
            centreRect1y = (i[1] + i[3])/2
            xaxis_person1.append(centreRect1x)
            yaxis_person1.append(centreRect1y)
        count = count + 1

```

```python
new_axis = []
new_yxis = []

newcount =0
for i,j in zip(xaxis_person1, yaxis_person1):
    if(newcount == 5):
        new_axis.append(i)
        new_yxis.append(j)
        newcount=0
    newcount = newcount + 1 


```
